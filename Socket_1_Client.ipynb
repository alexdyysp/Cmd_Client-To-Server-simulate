```python
#Client 
import socket
#声明SOCKET类型，同时生成连接对象
client = socket.socket()
client.connect(('localhost',6969))
#
print("try to connect server...")
while True:
    msg = input(">>:").strip()
    if len(msg) == 0 : continue  #加了这个，可以断开连接也能继续连上发消息
    client.send(msg.encode("utf-8"))
    data_cmd_size = client.recv(1024)  #先收size
    #data_cmd = client.recv(1024)  #再收数据
    print("from server:")
    recv_size = 0
    recv_data = b''
    while recv_size < int(data_cmd_size.decode()) :
        data = client.recv(1024)
        recv_size += len(data)  #
        recv_data += data
        #print(data.decode())
    else:
        print("cmd receive done....",recv_size)
        print(recv_data.decode())
#
client.close()
```
